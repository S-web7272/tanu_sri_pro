```python
df.rename({'OdName':'Country',
           'AreaName':'Continent',
           'RegName':'Region',
           'DevName':'Status'},axis=1, inplace=True)
df.head()
         
```

```python
df.set_index('Country',inplace=True)
```

```python
df.head()[2013].plot(kind='bar')
```

```python
cols = ['Type','Coverage','AREA','REG','DEV']
df.drop(cols,axis=1,inplace=True)
df.head()
```

```python
df.groupby('Continent')[2013].sum().plot(kind='pie',figsize=(8,8),title='continent wise immigration in 2013',autopct='%.3f%%')
plt.show()
```

```python
df.groupby('Continent')[1985,2013].sum().plot(kind='pie',subplots=True,figsize=(16,8),title='continent wise immigration in 1985,2013',autopct='%.3f%%')
plt.show()
```
